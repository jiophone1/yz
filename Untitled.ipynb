```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load grayscale image
gray = cv2.imread("sk.jpg", cv2.COLOR_RGB2YUV_YUY2)

# Step 2: Normalize to 0-1
gray = gray / 255.0

# Step 3: Fake color image creation (map gray to R, G, B with different weights)
r = gray * 1.0       # Keep red strong
g = gray * 0.8       # Slightly less green
b = gray * 0.6       # Even less blue

# Stack to form RGB image
colorized = np.stack((r, g, b), axis=2)

# Step 4: Show both
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')
plt.title("Original Grayscale")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(colorized)
plt.title("Simulated Colorized (Offline)")
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Step 1: Load grayscale image
img = cv2.imread('flower.jpg', cv2.bgsegm)

# Step 2: Apply colormap (pseudo-color)
color_img = cv2.applyColorMap(img, cv2.VIRIDIS)  # You can try COLORMAP_OCEAN, VIRIDIS etc.

# Step 3: Show results
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original B&W")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB))
plt.title("Fake Colorized")
plt.axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# -------- Step 1: Load grayscale image --------
gray = cv2.imread("flower.jpg", cv2.ADAPTIVE_THRESH_GAUSSIAN_C)
h, w = gray.shape

# -------- Step 2: Define control points --------
# 0  pure black
# mid_val  your chosen meeting point gray level (0255)
# 255  pure white

mid_val = 128
# Choose a real-like RGB for mid_val, e.g., a warm beige:
mid_rgb = np.array([220, 200, 180]) / 255.0  # normalized

# We'll map:
# gray=0    [0,0,0]
# gray=mid_val  mid_rgb
# gray=255  [1,1,1]

# -------- Step 3: Build lookup table --------
# Create an empty LUT of shape (256,3)
LUT = np.zeros((256, 3), dtype=np.float32)

# Segment 1: [0  mid_val]
for g in range(0, mid_val+1):
    t = g / mid_val  # 01
    LUT[g] = (1 - t) * np.array([0,0,0]) + t * mid_rgb

# Segment 2: [mid_val  255]
for g in range(mid_val, 256):
    t = (g - mid_val) / (255 - mid_val)  # 01
    LUT[g] = (1 - t) * mid_rgb + t * np.array([1,1,1])

# -------- Step 4: Apply LUT --------
# Flatten gray to map quickly
flat = gray.flatten()
color_flat = LUT[flat]          # shape (h*w,3)
colorized = color_flat.reshape((h, w, 3))

# -------- Step 5: Show result --------
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Original B&W")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(colorized)
plt.title(f"Controlled Color (mid={mid_val}{(mid_rgb*255).astype(int)})")
plt.axis('off')

plt.show()
```
